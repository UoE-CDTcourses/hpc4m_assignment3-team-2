```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

**Part I**

```python
Size = pd.read_csv('Test_size.txt', delim_whitespace=True, header=None)
```

```python
plt.plot(Size[0], Size[2])
plt.title('Size vs Time')
plt.xlabel('Message size (bytes)')
plt.ylabel('Time (s)')
plt.savefig('Fig-1_size-time.png', transparent = True, bbox_inches = 'tight')
plt.show()
```

```python
np.poly1d(np.polyfit(Size[0], Size[2], 12))(0.0)
```

```python
np.polyfit(Size[0], Size[2], 12, full=True)
```

```python
plt.plot(Size[0], Size[4])
plt.title('Size vs Bandwith')
plt.xlabel('Message size (bytes)')
plt.ylabel('Bandwith (MB/s)')
plt.savefig('Fig-2_time-bandwith.png', transparent = True, bbox_inches = 'tight')
plt.show()
```

**Part II**

```python
Its = pd.read_csv('Test_its.txt', delim_whitespace=True, header=None)
```

```python
plt.plot(Its[1], Its[2])
plt.title('Iterations vs Time')
plt.xlabel('Iterations size (bytes)')
plt.ylabel('Time (s)')
plt.savefig('Fig-3_its-time.png', transparent = True, bbox_inches = 'tight')
plt.show()
```
